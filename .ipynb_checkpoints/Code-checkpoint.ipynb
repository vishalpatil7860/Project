```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dot, Dense
from sklearn.metrics import mean_squared_error
from math import sqrt
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
import warnings

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('data.csv', encoding='ISO-8859-1')
data_info = data.info()
data_head = data.head()
data_info, data_head
```

```python
data.isnull().sum()
```

```python
# Cleaning the data

# # 1. Removing duplicates
# data_cleaned = data.drop_duplicates()

# 2. Handling missing values
data_cleaned = data.dropna(subset=['CustomerID', 'Description'])
# data_cleaned = data.dropna(subset=['Description'])

# Data Preprocessing
# Convert InvoiceDate to datetime
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

# Calculate TotalSpend
data['TotalSpend'] = data['Quantity'] * data['UnitPrice']

# Drop rows with missing CustomerID or Description
data_cleaned = data.dropna(subset=['CustomerID', 'Description'])
# 3. Ensuring consistent naming for products and countries
# Removing leading and trailing spaces from Description and Country
data_cleaned['Description'] = data_cleaned['Description'].str.strip()
data_cleaned['Country'] = data_cleaned['Country'].str.strip()

# 4. Dealing with outliers
# Removing entries with negative or zero quantities and prices, as they are likely to be returns or errors
data_cleaned = data_cleaned[(data_cleaned['Quantity'] > 0) & (data_cleaned['UnitPrice'] > 0)]

# cleaned_data_info

data_cleaned.isnull().sum()
```

```python
# Setting up visualization styles
sns.set(style="whitegrid")

# Analyzing the distribution of purchases across countries
country_distribution = data['Country'].value_counts()

# Display the top 10 countries in terms of number of transactions
plt.figure(figsize=(12, 6))
sns.barplot(x=country_distribution.head(10).values, y=country_distribution.head(10).index)
plt.title('Top 10 Countries by Number of Transactions')
plt.xlabel('Number of Transactions')
plt.ylabel('Country')
plt.show()

country_distribution.head(10)
```

```python
# Import necessary libraries
import numpy as np

# Convert InvoiceDate to datetime
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

# Calculating total spend per invoice
data['TotalSpend'] = data['Quantity'] * data['UnitPrice']

# Frequency of Purchases
purchase_frequency = data.groupby('CustomerID')['InvoiceDate'].nunique()

# Average Spend per Purchase, segmented by country
average_spend_per_country = data.groupby('Country')['TotalSpend'].mean()

# Product Popularity
popular_products = data['Description'].value_counts().head(10)

# Seasonal Trends
data['Month'] = data['InvoiceDate'].dt.month
seasonal_trends = data.groupby('Month')['InvoiceNo'].nunique()

# Printing the results
print("Frequency of Purchases (Top 10 Customers):")
print(purchase_frequency.sort_values(ascending=False).head(10))
print("\nAverage Spend per Purchase by Country:")
print(average_spend_per_country.sort_values(ascending=False).head(10))
print("\nMost Popular Products:")
print(popular_products)
print("\nSeasonal Trends (Number of Invoices per Month):")
print(seasonal_trends)
```

```python
# Create a user-item interaction matrix
interaction_matrix = data_cleaned.pivot_table(index='CustomerID', columns='Description', values='TotalSpend', fill_value=0)

# Flatten the matrix
interaction_df = interaction_matrix.reset_index().melt(id_vars='CustomerID', var_name='Description', value_name='TotalSpend')

# Encode CustomerID and Description
interaction_df['CustomerID'] = interaction_df['CustomerID'].astype("category").cat.codes
interaction_df['Description'] = interaction_df['Description'].astype("category").cat.codes

# Split the data
train_data, test_data = train_test_split(interaction_df, test_size=0.3, random_state=42)

# Model Building
user_input = Input(shape=(1,), name='user_input')
user_embedding = Embedding(input_dim=data_cleaned['CustomerID'].nunique(), output_dim=50, name='user_embedding')(user_input)
user_vec = Flatten(name='flatten_users')(user_embedding)

item_input = Input(shape=(1,), name='item_input')
item_embedding = Embedding(input_dim=data_cleaned['Description'].nunique(), output_dim=50, name='item_embedding')(item_input)
item_vec = Flatten(name='flatten_items')(item_embedding)

product = Dot(name='dot_product', axes=1)([user_vec, item_vec])
output = Dense(1, activation='relu')(product)

model = Model(inputs=[user_input, item_input], outputs=output)
model.compile(optimizer='adam', loss='mean_squared_error')

# Model Training
history = model.fit([train_data['CustomerID'], train_data['Description']], train_data['TotalSpend'], epochs=10, batch_size=256, verbose=1)

model.save('recommendation_model.h5')

# Model Evaluation
test_predictions = model.predict([test_data['CustomerID'], test_data['Description']])
rmse = sqrt(mean_squared_error(test_data['TotalSpend'], test_predictions))
print(f"Test RMSE: {rmse}")
```
